```python
import cv2
import mediapipe as mp
import numpy as np
from PIL import Image
import numpy
```

```python
class Pose_Detection:
    def __init__(self,path):
        self.mp_drawing = mp.solutions.drawing_utils
        self.mp_pose = mp.solutions.pose
        self.yoga_pose = cv2.imread(path)
        self.co_ordinates=[]
    def estimate_pos(self):
        with self.mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:

            image = cv2.cvtColor(self.yoga_pose, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False

            # Make detection
            results = pose.process(image)

            return results
    def fill_coordinates(results,self):
        for lndmrk,i in zip(self.mp_pose.PoseLandmark,list(results.pose_landmarks.landmark)):
            ind=str(lndmrk).index('.')
            points={
                "feature":str(lndmrk)[ind+1:],
                "x":i.x,
                "y":i.y,
                "z":i.z
            }
            self.co_ordinates.append(points)
    
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
yoga_pose = cv2.imread("pos.png")
co_ordinates=[]
```

```python
def estimate_pos(yoga_pose):
    with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:

        image = cv2.cvtColor(yoga_pose, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        # Make detection
        results = pose.process(image)
        
        return results
        
```

```python
results=estimate_pos(yoga_pose)
```

```python
def fill_coordinates(results):
    for lndmrk,i in zip(mp_pose.PoseLandmark,list(results.pose_landmarks.landmark)):
        ind=str(lndmrk).index('.')
        points={
            "feature":str(lndmrk)[ind+1:],
            "x":i.x,
            "y":i.y,
            "z":i.z,
            "As-Tuple":[i.x,i.y,i.z]
        }
        co_ordinates.append(points)
```

```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
```

```python
def determine_angles():
    angles_mapping={'Head':calculate_angle()
                    'Left-Elbow':calculate_angle()
                    'Right-Elbow',
                    'Left-Shoulder',
                    'Right-Shoulder',
                    'Left-Hip',
                    'Right-Hip',
                    'Left-Knee',
                    'Right-Knee',
                    'Left-Ankle',
                    'Right-Ankle'}
    
```
